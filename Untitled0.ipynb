```python
file_name = 'WEO_Data.csv'
df2 = pd.read_csv(file_name)
```

```python
# Filter the DataFrame for 'United States'
us_df = df2[df2['Country'] == 'United States']

# Melt the DataFrame to make the years columns into rows
melted_df = us_df.melt(id_vars=['Subject Descriptor'], value_vars=['1980', '1981', '1982', '1983', '1984',
                                                                   '1985', '1986', '1987', '1988', '1989',
                                                                   '1990', '1991', '1992', '1993', '1994',
                                                                   '1995', '1996', '1997', '1998', '1999',
                                                                   '2000', '2001', '2002', '2003', '2004',
                                                                   '2005', '2006', '2007', '2008', '2009',
                                                                   '2010', '2011', '2012', '2013', '2014',
                                                                   '2015', '2016', '2017', '2018', '2019',
                                                                   '2020', '2021', '2022', '2023'],
                  var_name='Year', value_name='Value')

# Pivot the melted DataFrame to get the desired shape
pivoted_df = melted_df.pivot_table(index='Year', columns='Subject Descriptor', values='Value', aggfunc='first')

# Reset the index if you want 'Year' to be a column again
pivoted_df.reset_index(inplace=True)
##pivoted_df['Year'] = pd.to_numeric(pivoted_df['Year'], errors='coerce')  # 'coerce' will set invalid parsing as NaN

##pivoted_df = pivoted_df[pivoted_df['Year'] > 1997]
##pivoted_df['Year'] = pd.to_numeric(pivoted_df['Year'], errors='coerce')

##pivoted_df['Current account balance'] = pd.to_numeric(pivoted_df['Current account balance'], errors='coerce')

##pivoted_df.dropna(subset=['Current account balance'], inplace=True)

##pivoted_df_filtered = pivoted_df[pivoted_df['Year'] > 1997]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  # Ensure pandas is imported

plt.figure(figsize=(10, 6))  # Set the figure size for better readability

# Convert 'Year' column to numeric if it's not already
pivoted_df['Year'] = pd.to_numeric(pivoted_df['Year'])

# Plot 'Gross national savings' and 'Total investment'
line1, = plt.plot(pivoted_df['Year'], pivoted_df['Gross national savings'], label='Gross National Savings', color='blue')
line2, = plt.plot(pivoted_df['Year'], pivoted_df['Total investment'], label='Total Investment', color='red')

# Add some plot decorations
plt.title('United States')
plt.ylabel('Percent of GDP')
plt.xlim(1980, 2023)  # Set x-axis limits
ticks = np.arange(1980, 2024, 5)  # Define custom ticks for every 5 years
plt.xticks(ticks, rotation=0)  # Set x-axis ticks
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Create a secondary y-axis for the 'Current account balance' bar chart
ax2 = plt.twinx()
bars = ax2.bar(pivoted_df['Year'], pivoted_df['Current account balance'], label='Current Account Balance', color='orange', alpha=0.4, width=0.9)
ax2.set_ylabel('Percent of GDP')
ax2.set_ylim(-8, 2)  # Set y-axis limits for the bar chart
ax2.axhline(y=0, color='black', linewidth=0.8)


# Gather all legends together below the plot without the surrounding box
handles = [line1, line2, bars]
labels = ['Savings (Left)', 'Investment (Left)', 'Current Account (Right)']
plt.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, frameon=False)

#plt.savefig('US.png', dpi=300, bbox_inches='tight')
plt.show()
#files.download('US.png')
```
