```python
df.fillna(0, inplace=True)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
# Filter the DataFrame for years 1836 to 1925
df_filtered = df[(df['year'] >= 1836) & (df['year'] <= 1925)]

# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Set the positions for the bars
bar_width = 0.35  # Width of the bars
index = df_filtered['year']  # X-axis values
index_shifted = index + bar_width  # Shifted positions for the second set of bars

# Plot the first set of bars for 'patentcity'
plt.bar(index, df_filtered['patentcity'], bar_width, label='Patent City', color='b')

# Plot the second set of bars for 'kelly', slightly offset to avoid overlap
plt.bar(index_shifted, df_filtered['kelly'], bar_width, label='Kelly', color='r')

# Add some labels and a legend
plt.xlabel('Year')
plt.ylabel('Values')
plt.title('Comparison of Patent City and Kelly over Years')
plt.xticks(index + bar_width / 2, df_filtered['year'].unique(), rotation=90)  # Set the position and labels of the ticks
plt.legend()

# Display the plot
plt.tight_layout()  # Adjust layout to not cut off labels
plt.show()

```

```python
df_filtered = df[(df['year'] >= 1836) & (df['year'] <= 1925)]

fig, ax = plt.subplots(figsize=(10, 6))

# Number of bars/groups
n_bars = len(df_filtered.index)

# Locations for the groups
index = np.linspace(1836, 1925, n_bars)  # Generate linearly spaced positions to match the range

# Width of each bar
bar_width = (index[1] - index[0]) * 0.25  # Calculate width based on spacing

# Plotting the bars
bars1 = ax.bar(index - bar_width / 2, df_filtered['kelly'], bar_width, color='blue', alpha=0.3, label='Kelly Dataset')
bars2 = ax.bar(index + bar_width / 2, df_filtered['patentcity'], bar_width, color='red', alpha=0.8, label='PatentCity')


#plt.title('Comparative Analysis of Patent Counts Pre-1925: PatentCity and Kelly Dataset', fontsize=12)
ax.set_ylabel('Patent counts')
# Setting x-ticks and labels for every 10 years from 1920 to 2020
tick_positions = range(1830, 1930, 10)
ax.set_xticks(tick_positions)
ax.set_xticklabels(tick_positions, rotation=45)  # Rotate x-axis labels for better readability

# Setting the x-axis limit from 1925 to 2017
plt.xlim(1836, 1925)

# Additional customizations
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.grid(axis='x', linestyle='--', alpha=0.5)
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2, frameon=False)

# Showing the plot
plt.tight_layout()  # Adjust layout to not cut off labels
#plt.savefig('before_1925.png', dpi=300)
plt.show()
#files.download('before_1925.png')
```
